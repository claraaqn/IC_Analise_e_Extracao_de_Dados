```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mbg = pd.read_csv('media_bandeira_gasolina.csv')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15, 6))
sns.lineplot(x='Ano', y='Valor de Venda', hue='Bandeira', data=mbg, palette="tab10", linewidth=0.5, style=True)
plt.xticks(range(2004, 2024), rotation=45, ha='right')
plt.yticks(range(2,8))
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```
